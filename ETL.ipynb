```python
import pandas as pd
```

# Extraction

```python
df_raw = pd.read_csv('data/wid_data_long_income.csv', skiprows=1, header=None, sep=';')
```

```python
tables = pd.read_html('https://wid.world/codes-dictionary/')
```

# Transformation

```python
df_raw.columns = ['country', 'indicator', 'percentile', 'year', 'sptinc']
```

```python
df_raw['WID code'] = [c.replace('\n',' ').split('_')[-1].split(' ')[0] for c in df_raw['indicator']]
```

```python
df_raw['share'] = [c.replace('\n', '|').split('|')[2] for c in df_raw['indicator']]
```

```python
df_raw
```

```python
df = tables[4][['WID code', 'region (continental)', 'region (political)', 'region (other)']]\
     .merge(df_raw[['WID code', 'country','share', 'sptinc']], how='left', on='WID code')\
     .dropna()
```

One way to prove that there are no repeated columns

```python
df.size == df.drop_duplicates().size
```
